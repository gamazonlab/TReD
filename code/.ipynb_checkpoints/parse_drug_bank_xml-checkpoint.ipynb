```python
import untangle# 导入该库用以将 XML 文件映射为一个 Python 对象，使相同名字的元素值组成数组
import pandas as pd# 导入 pandas 库用以存储结果数据
import zipfile
```

```python
##unzip file
def unzip_file(zip_path, extract_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)

# 指定zip文件路径和解压缩目标路径
zip_path = '/gpfs52/data/g_gamazon_lab/zhoud2/data/drugbank/raw/drugbank_all_full_database_5.1.10.xml.zip'
extract_path = '/gpfs52/data/g_gamazon_lab/zhoud2/data/drugbank/raw/extract/'

# 调用解压缩函数
unzip_file(zip_path, extract_path)
```

```python
# 读取 XML 文件
#drugbank_xml = untangle.parse("/gpfs52/data/g_gamazon_lab/zhoud2/data/drugbank/raw/extract/full database.xml")
df_drugbank_sm=pd.DataFrame(columns=['type',"drugbank_id","name","state",'target','cas','group','indication',"description"])
df_drugbank_sm
```

```python
i = 0
for drug in drugbank_xml.drugbank.drug:# 对每一个子元素进行循环
    drug_type = str(drug["type"])# 获取每个子元素的 "type" 属性
    df_drugbank_sm.loc[i, "type"] = drug_type
    
    #Get drugbank_id，获取 drug ID 属性
    for id in drug.drugbank_id:
    #由于 drugbank 中每个药物有多个不同的 ID，这里选取 primary ID
        if str(id["primary"])=="true":
            df_drugbank_sm.loc[i, "drugbank_id"]=id.cdata# 获取 ID 值，下同
    
    #Drug name，获取药物名称
    df_drugbank_sm.loc[i,"name"]=drug.name.cdata
    
    #state
    if 'state' in dir(drug):
        drug_state = drug.state.cdata
        df_drugbank_sm.loc[i,"state"] = drug_state
        
    #target
    df_drugbank_sm.loc[i,"target"] = drug.targets.cdata
        
    #Drug CAS，获取药物 CAS number
    df_drugbank_sm.loc[i, "cas"]=drug.cas_number.cdata

    #Drug group，获取药物所处的临床阶段
    groupcdata = ""
    for group in drug.groups.group:
        groupcdata = groupcdata+group.cdata+","
    df_drugbank_sm.loc[i, "group"]=groupcdata

    #Drug indication，获取药物适应症
    df_drugbank_sm.loc[i, "indication"]=drug.indication.cdata
    #state

    #Drug descriptions, 获取药物的描述信息
    df_drugbank_sm.loc[i, "description"] = drug.description.cdata
    
    i += 1
```
